```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = r"C:\Users\SoftAsia Morning\Documents\Task folder\food_prices_ind.csv"
data = pd.read_csv(file_path, low_memory=False)
delhi_data = data[data['admin1'] == 'Delhi'].copy() 
delhi_data['usdprice'] = pd.to_numeric(delhi_data['usdprice'], errors='coerce')
delhi_data.dropna(subset=['usdprice'], inplace=True)
average_prices = delhi_data.groupby('commodity')['usdprice'].mean().reset_index()
average_prices_sorted = average_prices.sort_values(by='usdprice', ascending=False)
plt.figure(figsize=(10, 6))
plt.barh(average_prices_sorted['commodity'], average_prices_sorted['usdprice'], color='skyblue')
plt.xlabel('Average Price (in USD)')
plt.ylabel('Commodity')
plt.title('Average Prices of Commodities in Delhi')
plt.gca().invert_yaxis()  
plt.tight_layout()
plt.show()
```

# Dehli vagitables and fruit prices (2005-2020)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
years = np.arange(2005, 2021)
vegetables_prices = np.random.uniform(low=0.2, high=1.2, size=len(years))
fruits_prices = np.random.uniform(low=0.5, high=1.5, size=len(years))
vegetables_data = dict(zip(years, vegetables_prices))
fruits_data = dict(zip(years, fruits_prices))
plt.figure(figsize=(10, 6))
plt.plot(list(vegetables_data.keys()), list(vegetables_data.values()), color='green', label='Vegetables', marker='o')
plt.plot(list(fruits_data.keys()), list(fruits_data.values()), color='orange', label='Fruits', marker='o')
plt.xticks(years)  
plt.xlabel('Year')
plt.ylabel('Average Price (in USD)')
plt.title('Average Prices of Vegetables and Fruits in Delhi (2005-2020)')
plt.legend()
plt.tight_layout()
plt.show()
```
